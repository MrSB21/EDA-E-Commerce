### Dealing with missing values (Mean/Mode/Median)

### Removing rows with negative values (Quantity column)

```python
sum(df['Quantity']<0)
```

```python
df[df.Quantity<0].index
```

```python
df.drop(index=df[df.Quantity<0].index,inplace=True)
df[df.Quantity<0]
```

### Adding Columns (Month, Day, Hour) for Invoice

```python
import datetime as dt

# Convert "InvoiceDate" column to datetime data type
df['InvoiceDate']=pd.to_datetime(df.InvoiceDate)

# Create another month column
df['Month']=df['InvoiceDate'].dt.month

# Create another day column
df['Day']=df['InvoiceDate'].dt.day

# Create another hour column
df['Hour']=df['InvoiceDate'].dt.hour
```
